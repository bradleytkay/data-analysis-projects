### Creating Visualizations

#### Line Chart

```python
# Single line chart

top_10 = df.head(10)

plt.plot(top_10['artist(s)_name'], top_10['streams'])
plt.xlabel('Artist(s)_Name')
plt.ylabel('Steams')
plt.xticks(rotation=90)
plt.show()

# This chart isn't making any sense to me... Shouldn't the line be undulating to keep track
# with the number of streams relative to other data points?  How do I get this chart to function
# the way I actually want it to... ?
```

```python
# Multiple Line chart 

top_10 = df.head(10)

plt.plot(top_10['artist(s)_name'], top_10['streams'], label = 'Streams')
plt.plot(top_10['artist(s)_name'], top_10['released_year'], label = 'Released Year')
plt.plot(top_10['artist(s)_name'], top_10['bpm'], label = 'BPM')
plt.xlabel('Artist(s)_Name')
plt.ylabel('')
plt.xticks(rotation=60)

# This is not working at all like I would expect or would want... I thought labels
# would... you know... LABEL each line.  The lines themselves also aren't doing what
# I'd expect.  I feel like I have a lot to learn about matplotlib visualizations...
```
